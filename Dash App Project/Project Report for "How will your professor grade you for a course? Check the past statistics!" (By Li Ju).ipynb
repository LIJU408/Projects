## Section 1: Dataset Description

The dataset is called "uiuc-gpa-dataset.csv". Its source is 

"https://github.com/wadefagen/datasets/blob/master/gpa/uiuc-gpa-dataset.csv". The dataset contains 15784 rows and 21

columns. Generally the dataset showed the number of each letter grade in each course in UIUC, from 2010 summer 

semester to 2020 spring semester. Each row in the dataset shows the semester when the course is taught, the course 

subject, number, title, the numbers of each letter grade, the primary instructor, etc.

Here are the detailed descriptions for 21 variables in the dataset:

- `Year`: which year the course was taught in
- `Term`: which term the course was taught in 
- `Year-Term`: which year and which term the course was taught in 
- `Subject`: which subject the course belonged to
- `Number`: what course number the course had
- `Course Title`: the name of the course
- `A+`: how many students gained 'A+' in the course
- `A`: how many students gained 'A' in the course
- `A-`: how many students gained 'A-' in the course
- `B+`: how many students gained 'B+' in the course
- `B`: how many students gained 'B' in the course
- `B-`: how many students gained 'B-' in the course
- `C+`: how many students gained 'C+' in the course
- `C`: how many students gained 'C' in the course
- `C-`: how many students gained 'C-' in the course
- `D+`: how many students gained 'D+' in the course
- `D`: how many students gained 'D' in the course
- `D-`: how many students gained 'D-' in the course
- `F`: how many students gained 'F' in the course
- `W`: how many students withdrew the course
- `Primary Instructor`: the instructor of the course


## Section 2: Question of Interest

The goal of my web application is to help students check how professors grade students for a certain course they are 

interested in. For example, Henry is considering to register for STAT100 for next semester but he is uncertain that he 

should choose which section by which professor. Henry cares his GPA and he hopes STAT100's professor does not have 

very harsh grading metric. That is the time Henry should try my web application. If he inputs 'STAT' and '100' in two 

input boxex, he will see a vivid visualization about the distribution of letter grades ,"A", "B", "C", "D" and "F", 

graded by all professors taught STAT100 in past 10 years. If he clicks on each bar in the graph, he can see how many 

students graded by a certain professor as well as the average GPA the professor gave to students.
